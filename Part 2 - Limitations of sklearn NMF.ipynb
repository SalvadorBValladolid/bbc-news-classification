## Part 1
Load the movie ratings data (as in the HW3-recommender-system) and use matrix factorization technique(s) and predict the missing ratings from the test data. Measure the RMSE. You should use sklearn library.

```python
import pandas as pd
import numpy as np
```

```python
MV_users = pd.read_csv('data/users.csv')
MV_movies = pd.read_csv('data/movies.csv')
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
```

Describe the data to remeber how it's composed.

```python
print('MV_users data:')
print(MV_users.head())
print('-'*100)
print('MV_movies data:')
print(MV_movies.head())
print('-'*100)
print('train data:')
print(train.head())
print('-'*100)
print('test data:')
print(test.head())
print('-'*100)
```

Now, I'll use the NMF to predict the ratings in the test set.

```python
from sklearn.decomposition import NMF
from sklearn.metrics import mean_squared_error
```

First, I'm building the user x movies rating matrix, this is the one for applying the NMF.

```python
user_movies_matrix = train.pivot(index='uID', columns='mID', values='rating').fillna(0)
```

Notice, that I'm filling the non rated movies with '3'.

```python
user_movies_matrix
```

As we can see, the size is 6040 x 3664, that's because there are 6040 users and 3664 unique movies in the train set, let's see.

```python
print('Unique users in train:')
print(len(train['uID'].unique()))
print('Unique movies in train:')
print(len(train['mID'].unique()))
```

```python
nmf = NMF(n_components=20)
user_matrix = nmf.fit_transform(user_movies_matrix)
movies_matrix = nmf.components_
```

```python
user_matrix.shape
```

```python
movies_matrix.shape
```

Now, let's build the matriz rating predictions based on the user_matrix and movies_matrix

```python
R_pred = np.dot(user_matrix, movies_matrix)
R_pred = pd.DataFrame(R_pred, index=user_movies_matrix.index, columns=user_movies_matrix.columns)
```

```python
R_pred
```

```python
def predict_rating(uID, mID):
    try:
        return R_pred.loc[uID, mID]
    except KeyError:
        return 3  # If the user or movie aren't in the train set, return 3
```

```python
test['pred_rating'] = test.apply(lambda x: predict_rating(x['uID'], x['mID']), axis=1)
```

Finally, calculate the RMSE

```python
rmse = mean_squared_error(test['rating'], test['pred_rating'], squared=False)
print("RMSE:", rmse)
```

I obtained a RMSE of: 1.107.

Discuss the results and why they did not work well compared to simple baseline or similarity-based methods we’ve done in Module 3. Can you suggest a way(s) to fix it?

The reasons on why the NMF did not that well are:
- The user-movie matrix is large, and NMF is not that optimized for such matrices, so learning latent representation it's more difficult.
- NMF requires n_components, if it's small, the representation is worse, but if it's too large it can overfit.
- Increase max_iterations in NMF.
- It doesn't use similarity as similirarity-based methods, because they use the genre movie information for example, while NMF doesn't.
- Cold start problem: Users with few ratings can be difficult for NMF.
- I fill not rated movies with 0's, which can affect the performance.

Ways to fix:
- Tune the n_components for NMF, in oder to select a balanced n.
- Fill not rated movies with 3's, this can improve.
- Add NMF with global baseline (mean rating) for rare users/movies.
